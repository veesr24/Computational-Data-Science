```python
Arrests_2021 = police_df["Arrest_Year"] == 2021
police_2021_df = police_df[Arrests_2021]
police_2021_df
```

```python
Arrest_column_names = list(police_2021_df.columns)
print(Arrest_column_names)
```

```python
Arrest_column_names = list(police_2021_df.columns)
print("Column names of police_2021_df:")
for column_name in Arrest_column_names:
    print(f"- {column_name}")
```

```python
police_2021_raceweapons = police_2021_df[['_id', 'Sex', 'Perceived_Race', 'SearchReason_PossessWeapons']]
police_2021_raceweapons
```

```python
sex_distribution = police_2021_raceweapons['Sex'].value_counts()
race_distribution = police_2021_raceweapons['Perceived_Race'].value_counts()
search_reason_distribution = police_2021_raceweapons['SearchReason_PossessWeapons'].value_counts()
police_2021_raceweapons.to_excel('police_2021_raceweapons_with_distributions.xlsx', index=False)
print("Sex Distribution:\n", sex_distribution)
print("\nPerceived Race Distribution:\n", race_distribution)
print("\nSearch Reason Distribution:\n", search_reason_distribution)
```
