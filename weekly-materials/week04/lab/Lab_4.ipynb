```python
import pandas as pd

# Assuming police_df is the DataFrame obtained from reading 'ArrestsStripSearches.csv'

# Selecting columns from police_df
police_2021_raceweapons = police_df[['_id', 'Sex', 'Perceived_Race', 'SearchReason_PossessWeapons']]

# Display the head of the new DataFrame in table format
print(police_2021_raceweapons.head().to_markdown(index=False))

# Examine the distribution of Sex in table format
print(police_2021_raceweapons['Sex'].value_counts().to_frame().to_markdown())

# Examine the distribution of Perceived_Race in table format
print(police_2021_raceweapons['Perceived_Race'].value_counts().to_frame().to_markdown())

# Examine the distribution of SearchReason_PossessWeapons in table format
print(police_2021_raceweapons['SearchReason_PossessWeapons'].value_counts().to_frame().to_markdown())
```

```python
# Assuming police_df is the DataFrame obtained from reading 'ArrestsStripSearches.csv'

# Selecting columns from police_df
police_2021_raceweapons = police_df[['_id', 'Sex', 'Perceived_Race', 'SearchReason_PossessWeapons']]

# Display the head of the new DataFrame in a table format
police_2021_raceweapons.head(25)
```

```python
# Computing the distribution of Sex and display in table format
sex_distribution = police_2021_raceweapons['Sex'].value_counts().to_frame().reset_index().rename(columns={'index': 'Sex', 'Sex': 'Count'})
print(sex_distribution.to_markdown(index=False))

# Computing the distribution of Perceived_Race and display in table format
race_distribution = police_2021_raceweapons['Perceived_Race'].value_counts().to_frame().reset_index().rename(columns={'index': 'Perceived_Race', 'Perceived_Race': 'Count'})
print(race_distribution.to_markdown(index=False))

# Computing the distribution of SearchReason_PossessWeapons and display in table format
reason_distribution = police_2021_raceweapons['SearchReason_PossessWeapons'].value_counts().to_frame().reset_index().rename(columns={'index': 'SearchReason_PossessWeapons', 'SearchReason_PossessWeapons': 'Count'})
print(reason_distribution.to_markdown(index=False))
```

```python
import pandas as pd

# Assuming police_df is the DataFrame obtained from reading 'ArrestsStripSearches.csv'

# Selecting columns from police_df
police_2021_raceweapons = police_df[['_id', 'Sex', 'Perceived_Race', 'SearchReason_PossessWeapons']]

# Display the head of the new DataFrame in table format
print(police_2021_raceweapons.head().to_markdown(index=False))

# Compute the distribution of Sex and display in table format
sex_distribution = police_2021_raceweapons['Sex'].value_counts().to_frame().reset_index().rename(columns={'index': 'Sex', 'Sex': 'Count'})
print(sex_distribution.to_markdown(index=False))

# Compute the distribution of Perceived_Race and display in table format
race_distribution = police_2021_raceweapons['Perceived_Race'].value_counts().to_frame().reset_index().rename(columns={'index': 'Perceived_Race', 'Perceived_Race': 'Count'})
print(race_distribution.to_markdown(index=False))

# Compute the distribution of SearchReason_PossessWeapons, including NaN values, and display in table format
reason_distribution = police_2021_raceweapons['SearchReason_PossessWeapons'].value_counts(dropna=False).to_frame().reset_index().rename(columns={'index': 'SearchReason_PossessWeapons', 'SearchReason_PossessWeapons': 'Count'})
print(reason_distribution.to_markdown(index=False))
```
