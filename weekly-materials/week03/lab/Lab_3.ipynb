# District Names Extraction

This code assumes that the variable `district_data` has been previously populated from a file. The purpose of this code snippet is to extract district names from each line in `district_data` and print them with a bold header. The process involves iterating through each line, splitting it into entries, assuming the district name is in the first position, and adding it to a list called `district_names`. Finally, the extracted district names are printed with a bold header stating "DISTRICT NAMES." No specific sorting is performed in this code snippet.


```python

district_names = []

# Iterating through each line in the district_data
for line in district_data:
    # Splitting each line into a list of entries using the default whitespace delimiter
    entries = line.split()
    
    # Extracting the name of the district and removing the leading/trailing whitespaces
    district_name = entries[0].strip()
    
    # Adding the district name to the list
    district_names.append(district_name)

# Sorting the district names alphabetically
district_names = sorted(district_names)

print("DISTRICT NAMES (Alphabetical Order):")
for name in district_names:
    print(name)

```

# Analysis of Electoral Districts in Toronto

In this comprehensive analysis, we aim to gain insights into Canada's electoral districts, specifically focusing on those that prominently feature the term "Toronto" in their names.

## Districts Extraction

To initiate the analysis, we meticulously extract pertinent details from the electoral district dataset. For each district, we split the data into distinct components—namely, the district name and its population. The primary criterion for inclusion is the presence of the term "Toronto" in the district name. If this condition is met, the district name and population are logged in separate lists for further examination.

## Districts Listing

The subsequent step involves presenting a detailed listing of districts that meet the "Toronto" criteria. The list is thoughtfully organized in alphabetical order, facilitating ease of reference. Importantly, each district entry includes information about its population. Enumerated rankings accompany the district names, offering readers a clear understanding of the relative positions of each district.

## Additional Information

Beyond the district listing, we delve into the statistical landscape of Toronto districts. A thorough calculation reveals the total number of districts considered in this analysis. Furthermore, we ascertain the cumulative population across all selected districts. To provide a more nuanced view, the average population of these Toronto districts is also computed.

## Districts Ranking

To unravel patterns within the population distribution, we introduce a ranking system. Districts are ordered based on their population from the highest to the lowest. This sorting mechanism, facilitated by a custom key, elucidates the variations in population sizes among the Toronto districts. The rankings offer valuable insights into the demographic composition of the electoral landscape.

This holistic approach to analyzing Toronto's electoral districts serves to provide a nuanced understanding of the geographical and population dynamics within this specific region.

```python
toronto_district_names = []
toronto_district_populations = []

for line in district_data:
    entries = line.split(",")
    
    # Extracting the name of the district from the first part after splitting by commas
    district_name = entries[0].strip()
    
    # Extracting the population of the district
    district_population = int(entries[1].strip())
    
    # To check if "Toronto" is in the district name
    if "Toronto" in district_name:
        toronto_district_names.append(district_name)
        toronto_district_populations.append(district_population)

# Printing the list of district names containing "Toronto" in the first part
print("District Names containing 'Toronto' in the first part (Alphabetical Order):")
for i, name in enumerate(sorted(toronto_district_names)):
    print(f"{i+1}. {name} - Population: {toronto_district_populations[toronto_district_names.index(name)]}")

# Calculating and printiing additional information
num_districts = len(toronto_district_names)
total_population = sum(toronto_district_populations)
average_population = total_population / num_districts

print(f"\nNumber of Districts: {num_districts}")
print(f"Total Population: {total_population}")
print(f"Average Population: {average_population:.2f}")

# Printing the list of district names ranked by population (highest to lowest)
print("\nDistrict Names ranked by Population (Highest to Lowest):")
ranked_districts = sorted(zip(toronto_district_names, toronto_district_populations), key=lambda x: x[1], reverse=True)
for i, (name, population) in enumerate(ranked_districts):
    print(f"{i+1}. {name} - Population: {population}")



```

Finally, compute a list of tne names of electoral districts that are at least 30 characters long.

```python
long_district_names = []

for district_name in district_names:
    # Splitting the district_name by ","
    name_parts = district_name.split(",")
    
    # Extracting the first part (name)
    name = name_parts[0].strip()
    
    if len(name) >= 30:
        long_district_names.append(name)

# Printing the list of district names with at least 30 characters
print("District Names with at least 30 characters:")
for name in long_district_names:
    print(name)
```
